```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## 2.5 영상 형태 변환하고 크기 축소하기

* 이미지 gray scale 변환   
$ I = round(0.229 \times R + 0.587 \times G + 0.114 \times B)$

```python
# cv2 이용한 변환
image = cv2.imread("./imgs/maplestory-256x256.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# 공식을 이용한 변환
def convert_gray(image):
    r_image = image[..., -1]
    g_image = image[..., 1]
    b_image = image[..., 0]
    
    gray = np.round(0.229 * r_image + 0.587 * g_image + 0.114 * b_image)
    return gray

numpy_gray = convert_gray(image)

plt.subplot(1, 3, 1)
plt.imshow(image[..., ::-1])
plt.subplot(1, 3, 2)
plt.imshow(gray, cmap="gray")
plt.subplot(1, 3, 3)
plt.imshow(numpy_gray, cmap="gray")
```

```python
# 문제 2
a = np.random.randint(1, 10, 10)
print(a)
## (1), (2)
print(a.min()) # 최대값 출력
print(a.max()) # 최소값 출력
print(a.argmin()) # 최소값 위치의 index 출력
print(a.argmax()) # 최대값 위치의 index 출력
print(a.mean()) # 평균값 출력
print(a.sum()) # 합 출력
print(a.cumsum()) # 누적 합 출력
print(a.prod()) # 요소들의 곱 출력
print(a.cumprod()) # 요소들의 누적 곱 출력
  
```
